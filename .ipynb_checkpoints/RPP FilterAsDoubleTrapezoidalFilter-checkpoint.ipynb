## RPP filter for the double trapezoidal filter

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

### Lets define the exponential decay function with random noise   

```python
baseline=0.
amp=5
noise_amp=0.
tau=1500
L=4000
#L=length
```

```python
x=np.arange(-L/2,L/2,1)
def expo(amp, x, tau):
    y=[0]*len(x)#x*0
    for i in range (1, len(x)): 
        if (x[i] > 0):
            y[i] = amp * np.exp(-x[i]/tau)
        else:
            y[i] = 0
    return y
```

```python
E = expo(amp, x, tau)
```

```python
N=[0.]*L
N=np.array(N)
for i in range (0,L):
    N[i]=random.uniform(-1,1)
N=N*noise_amp
#now baseline offset
baseline_c=[baseline]*L
baseline_c=np.array(baseline_c)
A = baseline_c+expo(amp, x, tau)
A=np.array(A)
#Hence the total signal
T=A+N
```

```python
plt.plot(T)
plt.grid()
```

Lets define most of the numpy array

```python
#Initialization
v_n=T
L0=300
L1=100
L2=600
L3=100
L4=300

v_n=np.array(v_n)
R0_L=[0]*L
R1_L=[0]*L
R2_L=[0]*L
R0_1_L=[0]*L
R1_1_L=[0]*L
R2_1_L=[0]*L

R0_L1=[0]*L
R1_L1=[0]*L
R2_L1=[0]*L
R0_1_L1=[0]*L
R1_1_L1=[0]*L
R2_1_L1=[0]*L

R0_L2=[0]*L
R1_L2=[0]*L
R2_L2=[0]*L
R0_1_L2=[0]*L
R1_1_L2=[0]*L
R2_1_L2=[0]*L

R0_L3=[0]*L
R1_L3=[0]*L
R2_L3=[0]*L
R0_1_L3=[0]*L
R1_1_L3=[0]*L
R1_1_L4=[0]*L
R2_1_L3=[0]*L
R2_1_L4=[0]*L

R0_0_L=[0]*L
R0_1_L=[0]*L
R0_2_L=[0]*L

R0_0_L1=[0]*L
R0_1_L1=[0]*L
R0_2_L1=[0]*L

R0_0_L2=[0]*L
R0_1_L2=[0]*L
R0_2_L2=[0]*L

R0_0_L3=[0]*L
R0_1_L3=[0]*L
R0_2_L3=[0]*L

R0_0_L4=[0]*L
R0_1_L4=[0]*L
R0_2_L4=[0]*L

R0_L4=[0]*L
R1_L4=[0]*L
R2_L4=[0]*L

D=[0]*L
D1=[0]*L
D2=[0]*L
D3=[0]*L
```

```python
for i in range (0, L):
    if i>L0:
        #1st one
        #R_0_L[i]
        R0_L[i]=v_n[i] - v_n[i-L0]
        R0_1_L[i]=R0_1_L[i-1]+R0_L[i]  #r0_L0
        
        R1_L[i]=R0_1_L[i] - L0*v_n[i-L0] 
        R1_1_L[i]=R1_1_L[i-1]+R1_L[i]#r1_L0
        
        R2_L[i]=R1_1_L[i] - (0.5*L0*(L0+1))*v_n[i-L0] 
        R2_1_L[i]=R2_1_L[i-1]+R2_L[i]#r2_L0 
        

for i in range (0, L):
    if i>L1:
        R0_L1[i]=v_n[i] - v_n[i-L1] 
        R0_1_L1[i]=R0_1_L1[i-1]+R0_L1[i] #r0_L1
        
        R1_L1[i]=R0_1_L1[i] - L1*v_n[i-L1] 
        R1_1_L1[i]=R1_1_L1[i-1]+R1_L1[i] #r1_L1
        
        R2_L1[i]=R1_1_L1[i] - (0.5*L1*(L1+1))*v_n[i-L1] 
        R2_1_L1[i]=R2_1_L1[i-1]+R2_L1[i] #r2_L1 
for i in range (0, L):
    if i>L2:
        R0_L2[i]=v_n[i] - v_n[i-L2] 
        R0_1_L2[i]=R0_1_L2[i-1]+R0_L2[i] #r0_L2
        
        R1_L2[i]=R0_1_L2[i] - L2*v_n[i-L2] 
        R1_1_L2[i]=R1_1_L2[i-1]+R1_L2[i] #r1_L2
        
        R2_L2[i]=R1_1_L2[i] - (0.5*L2*(L2+1))*v_n[i-L2]  
        R2_1_L2[i]=R2_1_L2[i-1]+R2_L2[i] #r2_L2
        
for i in range (0, L):
    if i>L3:
        R0_L3[i]=v_n[i] - v_n[i-L3] 
        R0_1_L3[i]=R0_1_L3[i-1]+R0_L3[i] #r0_L3
        
        R1_L3[i]=R0_1_L3[i] - L3*v_n[i-L3]
        R1_1_L3[i]=R1_1_L3[i-1]+R1_L3[i] #r1_L1
        
        R2_L3[i]=R1_1_L3[i] - (0.5*L3*(L3+1))*v_n[i-L3] 
        R2_1_L3[i]=R2_1_L3[i-1]+R2_L3[i] #r2_L1 

for i in range (0, L):
    if i>L4:
        R0_L4[i]=v_n[i] - v_n[i-L4] 
        R0_1_L4[i]=R0_1_L4[i-1]+R0_L4[i] #r0_L3
        
        R1_L4[i]=R0_1_L4[i] - L4*v_n[i-L4]
        R1_1_L4[i]=R1_1_L4[i-1]+R1_L4[i] #r1_L1
        
        R2_L4[i]=R1_1_L4[i] - (0.5*L4*(L4+1))*v_n[i-L4] 
        R2_1_L4[i]=R2_1_L4[i-1]+R2_L4[i] #r2_L1 
```

```python
DTF=[0]*L
DTF1=[0]*L
DTF2=[0]*L
DTF3=[0]*L
DTF4=[0]*L
DT=[0]*L
```

```python
##now do this for double trapezoidal filter:
for i in range (L0+L1+L2+L3+L4,L):
    DTF[i]=-3.33333333e-03*R0_1_L[i]-2.22222222e-06*R1_1_L[i]-3.60199702e-23*R2_1_L[i]
    DTF1[i]=-6.66666667e-04*R0_1_L1[i-L0]+2.14387710e-20*R1_1_L1[i-L0]-4.94728649e-22*R2_1_L1[i-L0]
    DTF2[i]=2.66666667e-03 *R0_1_L2[i-L0-L1]+2.22222222e-06*R1_1_L2[i-L0-L1]+4.13825086e-23*R2_1_L2[i-L0-L1]
    DTF3[i]=6.66666667e-04*R0_1_L3[i-L0-L1-L2]-2.14387710e-20*R1_1_L3[i-L0-L1-L2]+4.94728649e-22*R2_1_L3[i-L0-L1-L2]
    DTF4[i]=-2.66666667e-03*R0_1_L4[i-L0-L1-L2-L3]-2.22222222e-06*R1_1_L4[i-L0-L1-L2-L3]-2.96298297e-23*R2_1_L4[i-L0-L1-L2-L3]
    
    DT[i]=DTF[i]+DTF1[i]+DTF2[i]+DTF3[i]+DTF4[i]
```

```python
plt.plot(DT)
#plt.xlim(2400,2900)
#plt.ylim(4,6)
plt.grid()
```
