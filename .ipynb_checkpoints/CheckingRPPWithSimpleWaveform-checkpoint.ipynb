## Checking RPP filters with simple waveforms

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v_check=np.concatenate([np.full(200,0),np.full(200,1)])
v_check
```

```python
R0_0_L=[0]*400
R0_1_L=[0]*400
R0_2_L=[0]*400

R1_0_L=[0]*400
R1_1_L=[0]*400

R2_0_L=[0]*400
R2_1_L=[0]*400

R3_0_L=[0]*400
R3_1_L=[0]*400

L0=100
for i in range (1,400):
    if (i>L0):
        #1st term
        R0_0_L[i] = v_check[i] - v_check[i-L0]
        R0_1_L[i]=R0_1_L[i-1]+R0_0_L[i]
        
        #2nd term
        R1_0_L[i]=R0_1_L[i]-L0*v_check[i-L0]
        R1_1_L[i]=R1_1_L[i-1]+R1_0_L[i]
        
        #3rd term
        R2_0_L[i]=R1_1_L[i]-0.5*L0*(L0+1)*v_check[i-L0]
        R2_1_L[i]=R2_1_L[i-1]+R2_0_L[i]
        
        #4th term
        R3_0_L[i]=R2_1_L[i]-1/6*L0*(L0+1)*(L0+2)*v_check[i-L0]
        R3_1_L[i]=R3_1_L[i-1]+R3_0_L[i]
        
```

```python
plt.plot(R0_0_L)
```

```python
plt.plot(R0_1_L)
```

```python
plt.plot(R1_0_L)
#plt.xlim(140,155)
plt.grid()
```

```python
plt.plot(R1_1_L)
plt.grid()
```
