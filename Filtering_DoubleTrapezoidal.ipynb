```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
amp=10
tau=500
x=np.arange(-3500,3500,1)
```

```python
def expo(amp, x, tau):
    y=[0]*len(x)
    for i in range (1, len(x)): 
        if (x[i] > 0):
            y[i] = amp * np.exp(-x[i]/tau)
        else:
            y[i] = 0
    return y
```

```python
A = expo(amp, x, tau)
plt.plot(A)
```

```python
#def trapFilter_new(input1, rise_time, flat_top, length, tau):
length=7000
rise_time=50
flat_top=100
input1=A

#initialize the arrays
trap = [0] *length
trap_o = [0]*length

#make a loop to obtain filtered signal
for i in range(0, length):
        #single trapezoidal filetr for exponential decay
        if i>=(4*rise_time +2*flat_top):
            trap[i] = trap[i-1] + input1[i]-input1[i-rise_time]*np.exp(-rise_time/tau)-input1[i-rise_time-flat_top] + input1[i-2*rise_time-flat_top]*np.exp(-1*rise_time/tau) -input1[i-2*rise_time-flat_top] + input1[i-3*rise_time-flat_top]*np.exp(-rise_time/tau) + input1[i-3*rise_time-2*flat_top] - input1[i-4*rise_time-2*flat_top]*np.exp(-rise_time/tau)  
            trap_o[i] = trap_o[i-1]-input1[i]+input1[i-rise_time]*np.exp(-rise_time/tau)+input1[i-rise_time-flat_top]-input1[i-2*rise_time-flat_top]*np.exp(-1*rise_time/tau) +input1[i-2*rise_time-flat_top] - input1[i-3*rise_time-flat_top]*np.exp(-rise_time/tau) - input1[i-3*rise_time-2*flat_top] + input1[i-4*rise_time-2*flat_top]*np.exp(-rise_time/tau) 
#now scale the output
for i in range(0, length):
    trap[i] = trap[i]/rise_time
    trap_o[i]=trap_o[i]/rise_time
```

```python
plt.plot(trap, color='r', label='DTF', zorder=2)
plt.plot(trap_o, color='g', label='DTF inverted', zorder=1)
plt.plot(A, color='b', label='wave', zorder=0)
plt.legend()
#plt.xlim(3900,5000)
plt.grid()
plt.show()
```
